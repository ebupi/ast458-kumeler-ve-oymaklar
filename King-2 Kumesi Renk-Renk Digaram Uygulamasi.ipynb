```python
# gerekli kutuphanelerin dahil edilmesi
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# verilerin dahil edilmesi
fileking = pd.read_csv("king2.csv") # king verisini pandas ile oku
filezams = pd.read_csv("SUNG_ZAMS.csv") # zams verisini pandas ile oku
dataking = pd.DataFrame(fileking) # verileri dataframe donustur
datazams = pd.DataFrame(filezams)
dataking.head()
```

```python
# verileri gorsellestirelim
plt.figure(figsize=(10,6))
plt.plot(dataking["B-V"],dataking["U-B"],".")
plt.title("King 2 Kümesi Renk-Renk Diagramı")
plt.xlabel("B-V")
plt.ylabel("U-B")
plt.gca().invert_yaxis() # y eksenini ters çevirdik
#plt.grid()
plt.show()
```

```python
# sung teorik verisini ekleyip tekrar cizdirelim
plt.figure(figsize=(10,6))
plt.plot(dataking["B-V"],dataking["U-B"],".",label="King 2 Yıldız Kümesi")
plt.plot(datazams["B-V"],datazams["U-B"],"-.",color="red",label="Sung Teorik")
plt.title("King 2 Kümesi & Sung Teorik \n Renk-Renk Diagramı")
plt.xlabel("B-V")
plt.ylabel("U-B")
plt.legend()
plt.gca().invert_yaxis()
#plt.grid()
plt.show()
```

```python
# pandas verilerinin numpy verilerine donusturulmesi
B_V0 = np.array(datazams["B-V"]) # pandas verisini numpy array cevirme
U_B0 = np.array(datazams["U-B"]) # pandas verisini numpy arary cevirme

EB_V = 0.31 # king 2 reddening(kizarma) = 0.31

# zams B-V arrayindeki(sutunundaki) her bir veriye EB_V kizarma degerini ekleyelim
B_V = np.array([i+EB_V for i in B_V0]) # kizartilmis B-V

# zams U-B arrayindeki(sutunundaki) her bir veriyi +0.72*kizarma+0.05*(kizarma**2)
# islemini uygulayalim
U_B = np.array([i+(0.72*EB_V)+0.05*(EB_V**2) for i in U_B0]) # kizartilmis U-B
```

```python
len(datazams["B-V"])
```

```python
# Teorik Sung, Kizartilmis Sung ve King2 kumesinin gorsellestirilmesi
plt.figure(figsize=(8,6))
plt.plot(dataking["B-V"],dataking["U-B"],".",label="King 2 Yıldız Kümesi") # king verisini ciz
plt.plot(B_V,U_B,"-.",label="Kizartilmis Sung",color="purple")             # kizartilmis sung veresini ciz
plt.plot(B_V0,datazams["U-B"],"-.",color="red",label="Sung Teorik") # zams verisinden sung verisini ciz
plt.title("King 2 Kümesi & Sung Teorik \n Renk-Renk Diagramı") # grafige isim ver
plt.xlabel("B-V") # x exeni isim ver
plt.ylabel("U-B") # y ekesenine isim ver

plt.gca().invert_yaxis() # y eksen degerlerini ters cevir

# kizartilmis sung ile teorik sung arasidaki egimin hesabi
x = [B_V0.min(),B_V.min()]
y = [U_B0.min(),U_B.min()]
egim = (y[0]-y[1])/x[0]-x[1]

print("Eğim : {:.4f}".format(egim))

plt.annotate('egim : {:.2f}'.format(egim), xy=(B_V0[0], U_B0[0]) ,xytext=(B_V[0],U_B[0]),
             arrowprops=dict(arrowstyle="->",facecolor='black',relpos=(B_V[0],U_B[0]+0.8)),xycoords="data")

plt.legend() # grafiklerin hangi veriye ait olduklarini göster
plt.show()
```

```python
x = [B_V0.min(),B_V.min()]
y = [U_B0.min(),U_B.min()]
egim = (y[0]-y[1])/x[0]-x[1]
print("Eğim : {:.2f}".format(egim))
gercek = 0.72
yaklasim = egim
yhata = (gercek-yaklasim)/gercek
yuzhata = yhata*100
print("Bagil yüzde hata % {:.2f}".format(yuzhata))
```

```python
#!pip install emcee
import emcee
import numpy as np
import matplotlib.pyplot as plt

# Örnek veri kümeleri
observed_bv = np.array([0.5, 1.2, 0.8, 0.9, 1.5])
template_bv = np.array([0.3, 0.8, 0.6, 1.0, 1.2])

# Log olasılık fonksiyonu
def ln_likelihood(theta, observed):
    mu, sigma = theta
    residuals = observed - mu
    return -0.5 * np.sum(residuals**2 / sigma**2 + np.log(2 * np.pi * sigma**2))

def ln_prior(theta):
    mu, sigma = theta
    if 0 < sigma < 10:
        return 0.0
    return -np.inf

def ln_posterior(theta, observed):
    ln_prior_val = ln_prior(theta)
    if not np.isfinite(ln_prior_val):
        return -np.inf
    return ln_prior_val + ln_likelihood(theta, observed)

# MCMC çalıştırma
ndim = 2
nwalkers = 100
nsteps = 2000

# Başlangıç noktalarını rastgele seçme
p0 = np.random.rand(nwalkers, ndim) * [10, 10]

# Hesaplamayı yürütme
sampler = emcee.EnsembleSampler(nwalkers, ndim, ln_posterior, args=[observed_bv])
sampler.run_mcmc(p0, nsteps)

# Sonuçları alıp görselleştirme
samples = sampler.get_chain(discard=100, flat=True)
mu_median = np.median(samples[:, 0])
sigma_median = np.median(samples[:, 1])
reddening = mu_median - template_bv

# Traceplot
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(samples[:, 0], color='b', alpha=0.3)
plt.axhline(mu_median, color='b', linestyle='--', label='Median')
plt.xlabel('Step')
plt.ylabel('$\mu$')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(samples[:, 1], color='r', alpha=0.3)
plt.axhline(sigma_median, color='r', linestyle='--', label='Median')
plt.xlabel('Step')
plt.ylabel('$\sigma$')
plt.legend()

plt.tight_layout()
plt.show()

# Kızıllaştırma hesaplama
print("Kızıllaştırma değeri: ", reddening)
```

```python
import numpy as np
import emcee

# Gözlemsel B-V verileri
observed_bv = np.array([0.5, 0.8, 1.2, 0.9])

# Zams B-V verileri
zams_bv = np.array([0.4, 0.7, 1.0, 1.1])

# Model fonksiyonu
def model(params):
    # Modelin hesaplanması
    kizillastirma = params[0]
    model_bv = zams_bv + kizillastirma
    return model_bv

# Log-likelihood fonksiyonu
def log_likelihood(params):
    model_bv = model(params)
    residual = observed_bv - model_bv
    chi_squared = np.sum(residual**2)
    return -0.5 * chi_squared

# Log-prior fonksiyonu
def log_prior(params):
    kizillastirma = params[0]
    if -1.0 < kizillastirma < 1.0:
        return 0.0
    return -np.inf

# Log-posterior fonksiyonu
def log_posterior(params):
    log_prior_value = log_prior(params)
    if not np.isfinite(log_prior_value):
        return -np.inf
    return log_prior_value + log_likelihood(params)

# Başlangıç değerleri
initial_params = [0.0]  # Kızıllaştırma başlangıç değeri

# Adım sayısı
nwalkers = 100
ndim = len(initial_params)
nsteps = 1000

# Walkerların başlangıç noktalarının belirlenmesi
p0 = np.random.randn(nwalkers, ndim) * 0.01 + initial_params

# Emcee örnekleme
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_posterior)
sampler.run_mcmc(p0, nsteps)

# Zincir çıktısı
samples = sampler.chain[:, :, :].reshape(-1, ndim)

# En iyi eşleştirme
best_params = samples[np.argmax(sampler.lnprobability)]

# Kızıllaştırma değeri
best_kizillastirma = best_params[0]

print("En iyi kızıllaştırma değeri:", best_kizillastirma)
```
